```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
df=pd.read_csv('weather_data.csv',parse_dates=['Formatted Date'])
df.dtypes
```

## Cleaning the dataset


```python
print(df.shape)
print(df.isnull().sum())
df.dtypes
```

### in this data set thier is some useless coloumns drop them

```python
df=df.drop(columns=['Loud Cover','Precip Type','Daily Summary'])
df
```

#### addiing new columns to data from foramtted date to know year wies,month,week,day,hour temp

```python
df['month']=df['Formatted Date'].dt.month
df['year']=df['Formatted Date'].dt.year
df['hour']=df['Formatted Date'].dt.hour
df['week']=df['Formatted Date'].dt.dayofweek
df['day']=df['Formatted Date'].dt.day
df
```

## EDA report 
#### 1.Uni_variate distribution
#### 2.Bi_variate distribution
#### 3.Multi_variate distribution

### Uni_variate distibution for both temp and  next apparent temperature


```python
plt.hist(df['Temperature (C)'])
```

```python
sns.distplot(df['Apparent Temperature (C)'],kde=True)
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(df['Temperature (C)'],shade=True,bw=10)
```

### Bi_variate distribution

```python

plt.plot(df['Humidity'],df['Temperature (C)'],'b*')
plt.axis(xmin=0,xmax=1.5,ymin=-20,ymax=40)
```

### Monthly Average Temperature

```python
month=df.groupby('month')['Temperature (C)'].mean()

plt.plot(month,'r*',month,'b--')
plt.axis(xmin=0,xmax=13)
plt.title('Monthly Average Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature')
```

### Monthly Average Apparent Temperature(C) using boxplot

```python
plt.figure(figsize=(15,8))
sns.boxplot(x='month',y='Apparent Temperature (C)',data=df)
plt.xlabel('Month')
plt.ylabel('Apparent Temperature(C)')
plt.title('Monthly Average  Apparent Temperature')
```

## kde(kernel density estimation)

```python
plt.figure(figsize=(10,8))
sns.kdeplot(df['Humidity'],df['Temperature (C)'],shade=True)
```

#### Hours Average Temperature Throught The Year

```python
#hour=df.groupby(['hour','month'])['Temperature (C)'].mean().plot(kind='bar',figsize=(10,8))
sns.barplot(x='hour',y='Temperature (C)',data=df)
plt.title('24_hours Average Temperatur from 2005-12:2007-01')
plt.xlabel('24_hours')
plt.ylabel('temperature')
```

# Jointplot

```python
sns.jointplot(df['Humidity'],df['Temperature (C)'],kind='kde')
```

```python

sns.jointplot(df['Humidity'],df['Apparent Temperature (C)'],kind='kde')
```

## Monthly Wise Average Temperature based caterogical data

```python

m=sns.FacetGrid(data=df,col='month',col_wrap=4,hue='Summary',palette=sns.light_palette('green',10),height=5,aspect=.5)
m=m.map(plt.scatter,'Humidity','Temperature (C)',alpha=0.9,s=100)
m.add_legend()
```

## Monthly Average Apparent Temperature based on caterogical data

```python
plt.figure(figsize=(11,8))
sns.lmplot(x='Humidity',y='Apparent Temperature (C)',data=df,hue='Summary',col='month',col_wrap=4,markers="+",aspect=0.5)
```
